```python
#contar informações
prateleira=['Banana','Maça','Laranja','uva']

#criar um loop que pergunte ao usuario se deseja adicionar uma fruta
# a prateleira só pode ter 7 produtos
# sempre que for adicionado um produto , mostrar
#dicas usar o while e append

print("Quantidade inicial:", len(prateleira))

# loop
while len(prateleira) < 7:
    resposta = input("Deseja adicionar uma fruta? (sim/nao): ").lower()

    if resposta == 'sim':
        fruta = input("Digite o nome da fruta: ")
        prateleira.append(fruta)

        #print("Fruta adicionada!")
        #print("Prateleira:", prateleira)
        #print("Quantidade:", len(prateleira))
   # elif resposta == 'nao':
    #    print("Encerrando...")
     #   break
    ##
    for i in prateleira:
        print(f'temos a fruta{i}')

    else:
        print("Resposta inválida! Digite 'sim' ou 'nao '.")

# mensagem final
print("Prateleira final:", prateleira)


```

```python
#usando o for
#criar um programa que pergunte ao usuario o nome de 3 musicas e adicionar em uma tebela
#mostrr todas elas da seguinte maneira
# a musica  1 é nome_da_Musica
#lista_musica =[]
print('='*60)
print('Bem vindo ao programa de lista de musica ')
print('='*60)
nome=input("digite seu nome ")

lista_musica = []

# loop das 3 músicas
for i in range(3): # range
    musica = input(f"Digite o nome da música {i+1}: ")
    lista_musica.append(musica)

print("\nLista de músicas do : ", nome)

# mostrar n
for i in range(len(lista_musica)):
    print(f"A música {i+1} é {lista_musica[i]} ")

```
