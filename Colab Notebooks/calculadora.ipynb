```python
# sem usar o numpy
# matemática básica
numero1 = float(input('informe o primeiro número'))
numero2 = float(input('informe o segundo número'))

# soma
soma = numero1 + numero2
print(soma)
# subtração
subtracao = numero1 - numero2
print(subtracao)
# divisão
divi = numero1/numero2
print(divi)
#multiplicação
multi = numero1 * numero2
print(multi)
```

```python
#usando numpy
import numpy as np

numero1 = float(input('informe o primeiro número'))
numero2 = float(input('informe o segundo número'))

#soma
soma =(f'soma:{np.add(numero1,numero2)}')
#sub
print(f'sub:{np.subtract(numero1,numero2)}')
#div
print(f'div:{np.divide(numero1,numero2)}')
#multi
print(f'multi:{np.multiply(numero1,numero2)}')
print(soma)
```
