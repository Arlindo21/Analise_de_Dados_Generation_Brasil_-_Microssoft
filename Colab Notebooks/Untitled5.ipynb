```python
import requests

# API clima
url_clima = 'https://api.open-meteo.com/v1/forecast'

# API geocodificação reversa (cidade)
url_geo = 'https://nominatim.openstreetmap.org/reverse'

latitude = float(input('Informe a latitude desejada: '))
longitude = float(input('Informe a longitude desejada: '))

# parâmetros clima
parametros_clima = {
    "latitude": latitude,
    "longitude": longitude,
    "current_weather": True,
}

# parâmetros localização
parametros_geo = {
    "lat": latitude,
    "lon": longitude,
    "format": "json"
}

# requisição clima
resposta_clima = requests.get(url_clima, params=parametros_clima)

# requisição localização (IMPORTANTE: header user-agent)
resposta_geo = requests.get(
    url_geo,
    params=parametros_geo,
    headers={"User-Agent": "meu_app_teste"}
)

if resposta_clima.status_code == 200 and resposta_geo.status_code == 200:

    dados_clima = resposta_clima.json()
    dados_geo = resposta_geo.json()

    temperatura = dados_clima['current_weather']['temperature']
    vento = dados_clima['current_weather']['windspeed']

    # pegando cidade (pode variar dependendo do local)
    endereco = dados_geo.get('address', {})
    cidade = endereco.get('city') or endereco.get('town') or endereco.get('village') or "Local desconhecido"
    pais = endereco.get('country', '')

    print(f'Local: {cidade} - {pais}')
    print(f'Temperatura: {temperatura} ºC')
    print(f'Velocidade do vento: {vento} km/h')

else:
    print('Erro na requisição')
```
