```python
#maneira 2 - importação via Raw 
#importação unitaria de um arquivo específico
#url do raw do arquivo
arquivo_1 = 'https://raw.githubusercontent.com/ChiaviniK/import-de-dados-com-python/refs/heads/main/example_data.csv'
arquivo_2 = 'https://raw.githubusercontent.com/ChiaviniK/import-de-dados-com-python/refs/heads/main/example_data.json'
arquivo_3 = 'https://github.com/ChiaviniK/import-de-dados-com-python/raw/refs/heads/main/example_data.xlsx'
arquivo_4 = 'https://raw.githubusercontent.com/ChiaviniK/import-de-dados-com-python/refs/heads/main/sample_data.csv'
#para importar os arquivos, basta usar a função de leitura do pandas
import pandas as pd
#csv 
df_csv_example= pd.read_csv(arquivo_1)
display(df_csv_example)
#json
df_json = pd.read_json(arquivo_2)
display(df_json)
#xlsx
#sempre que for um excel devo usar o engine 'openpyxl' para evitar erros de leitura
df_xlsx = pd.read_excel(arquivo_3, engine='openpyxl')
display(df_xlsx)
#csv2
df_csv2_sample = pd.read_csv(arquivo_4)
display(df_csv2_sample)
```
