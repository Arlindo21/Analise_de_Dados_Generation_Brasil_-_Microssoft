```python
jogador1 = input(' informe nome jogador 1 ')
jogador2 = input(' informe nome jogador 2 ')

pontuacao1 = 0
pontuacao2 = 0

movimento1 = None
movimento2 = None
```

```python
def escolher_jogada():
  # função
  print('='*20)
  print('as jogadas que temos são:')
  print('Pedra')
  print('Papel')
  print('Tesoura')
  print('='*20)
  print('DIGITE O TEXTO DA OPÇÃO')

  movimento1 = input(f'{jogador1}, informe sua jogada').lower()
  movimento2 = input(f'{jogador2}, informe sua jogada').lower()

  return movimento1, movimento2

def jogar():
  print(f'{jogador1} escolheu {movimento1}')
  print(f'{jogador2} escolheu {movimento2}')

  if movimento1 == "pedra" and movimento2 == 'papel':
    print(f"{jogador2} ganhou!")
    pontuacao2 += 1

  elif movimento1 == 'pedra' and movimento2 == 'tesoura':
    print(f'{jogador1} ganhou!')
    pontuacao1 += 1

  elif movimento1 == 'papel' and movimento2 == 'pedra':
    print(f'{jogador1} ganhou!')
    pontuacao1 += 1
  elif movimento1 == 'papel' and movimento2 == 'tesoura':
    print(f"{jogador2} ganhou!")
    pontuacao2 += 1
  elif movimento1 == 'tesoura' and movimento2 == 'pedra':
    print(f"{jogador2} ganhou!")
    pontuacao2 += 1
  elif movimento1 == 'tesoura' and movimento2 == 'papel':
    print(f'{jogador1} ganhou!')
    pontuacao1 += 1
  else:
    print('empatou')
def regras():
  print('pedra ganha de tesoura')
  print('papel ganha de pedra')
  print('tesoura ganha de papel')
```

```python
print('vamos iniciar o jogo')
print('1 -regras')
print('2 - jogar')
escolha =input('qual opção desejada? 1 ou 2')
if escolha == '1':
  regras()
elif escolha == '2':
  movimento1, movimento2 = escolher_jogada()
  jogar()
else:
  print('opção invalida')
```
