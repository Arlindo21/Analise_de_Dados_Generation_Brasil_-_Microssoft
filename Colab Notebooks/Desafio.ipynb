```python
import pandas as pd
```

```python
janeiro=pd.read_csv('/content/vendas_janeiro.csv')
fevereiro=pd.read_csv('/content/vendas_fevereiro.csv')
marco=pd.read_csv('/content/vendas_marco.csv')
```

```python
df_janeiro= pd.DataFrame(janeiro)
de_fevereiro= pd.DataFrame(fevereiro)
df_marco= pd.DataFrame(marco)
```

```python
df_tri_complto= pd.concat([df_janeiro,de_fevereiro,df_marco])
```

```python
df_tri_complto.to_csv('vendas_tri_complto.csv')
#
```
